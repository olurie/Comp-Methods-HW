```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
gamma = 0.5
beta = 0.975
r = 0.02
w = 1.0

# Consumption growth factor from Euler equation
growth_factor = (beta * (1 + r)) ** gamma
print(f"Consumption growth factor = {growth_factor:.6f}")

# Compute c0 using the present value budget constraint
numerator = w / r
denominator = 1 - (growth_factor / (1 + r))
c0 = numerator * denominator
print(f"Initial consumption c0 = {c0:.6f}")

# Generate consumption path over time
T = 50
t_grid = np.arange(T)
c_path = c0 * (growth_factor ** t_grid)


utility = (c_path ** (1-1/gamma))/(1-1/gamma)
utility_path = np.zeros(T)
for i in range(T):
    utility_path[i] = c_path[i] ** (1 - 1/gamma) / (1 - 1/gamma)

# Plot consumption path
plt.figure(figsize=(8, 5))
plt.plot(t_grid, c_path, marker='o')
plt.xlabel("Time t")
plt.ylabel("Consumption c_t")
plt.title("Optimal Consumption Path (Analytical Solution)")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(t_grid, utility_path, marker='o')
plt.xlabel("Time t")
plt.ylabel("Utility")
plt.title("Utility Path (Analytical Solution)")
plt.grid(True)
plt.show()
```
